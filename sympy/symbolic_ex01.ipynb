```python
from sympy import * 
x = Symbol('x')
y - Symbol('y')

facts = Q.positive(x), Q.positive(y)
with assuming(*facts):
    print(ask(Q.positive(2*x+y)))
```

```python
from sympy import Q, ask, MatrixSymbol
X = MatrixSymbol('X',2,2)
Y = MatrixSymbol('Y',2,3)
Z = MatrixSymbol('Z',2,2)
```

```python
ask(Q.symmetric(X*Z), Q.symmetric(X) & Q.symmetric(Z))
```

```python
ask(Q.symmetric(Y))
```

```python
print(ask(Q.symmetric(X)))
```

```python
facts = Q.symmetric(X), Q.symmetric(Z)
with assuming(*facts):
    print(ask(Q.symmetric(X*Z)))
```

```python
from sympy import euler_equations, Symbol, Function
x = Function('x')
t = Symbol('t')
```
