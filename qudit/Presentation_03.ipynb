```python
import numpy as np
from scipy.linalg import expm, sinm, cosm
```

```python
s0 = np.matrix([[1+0*1j,0],[0,1+0*1j]])
s1 = np.matrix([[0+0*1j,1+0*1j],[1,0]])
s2 = np.matrix([[0,-1j],[1j,0]])
s3 = np.matrix([[1+0*1j,0],[0,-1]])
```

```python
print(np.matmul(s1,s1))
```

```python
print(np.matmul(s2,s2))
```

```python
print(np.matmul(s3,s3))
```

```python
A = expm(1j*np.matmul(s1,s1))
```

```python
print('s0*s0 = ',s0**2)
print('s1*s1 = ',s1**2)
print('s2*s2 = ',s2**2)
print('s3*s3 = ',s3**2)
```
