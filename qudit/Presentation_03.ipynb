```python
import numpy as np

def Xdj_alpha(d,j,alpha):
    X = np.zeros((d,d), dtype = 'complex_')
    a = alpha[j];
    sn1 = 0;
    sn = 0;
    tmp = 0;
    for iter in range(j):
        tmp = tmp + alpha[iter]*np.conjugate(alpha[iter])
    sn1 = np.sqrt(tmp)
    sn = np.sqrt(tmp + alpha[j]*np.conjugate(alpha[j]))
    for iter in range(d):
        X[iter,iter]=1.0

    if sn > 0:
        X[j-1,j-1]=a/sn
        X[j-1,j]=-sn1/sn
        X[j,j-1]=np.conjugate(sn1/sn)
        X[j,j]=np.conjugate(a)/sn
    else :
        X[j-1,j-1]=1
        X[j-1,j]=0
        X[j,j-1]=0
        X[j,j]=1
    return X;
```

```python
import numpy as np

def Ydj_alpha(d,j,alpha):
    Y = np.zeros((d,d), dtype = 'complex_')
    a = alpha[j];
    sn1 = 0;
    sn = 0;
    tmp = 0;
    for iter in range(j):
        tmp = tmp + alpha[iter]*np.conjugate(alpha[iter])
    sn1 = np.sqrt(tmp)
    sn = np.sqrt(tmp + alpha[j]*np.conjugate(alpha[j]))
    for iter in range(d):
        Y[iter,iter]=1.0

    if sn > 0 and sn1 >0:
        Y[j-1,j-1]=np.conjugate(a/(sn*sn))
        Y[j-1,j]=sn1/(sn*sn)
        Y[j,j-1]=-np.conjugate(sn1/(sn*sn))
        Y[j,j]=a/(sn*sn)
    else :
        Y[j-1,j-1]=1
        Y[j-1,j]=0
        Y[j,j-1]=0
        Y[j,j]=1
    return Y;
```

```python
import numpy as np

def ketdj(d,j):
    ket = np.zeros([d,1], dtype='complex_')
    ket[j]=1.0+0*1j
    return ket;
```

```python
import numpy as np

def UA_cal(d,alpha):
    UA = np.identity(d, dtype = 'complex_')
    for iter in range(d-1):
        X = Xdj_alpha(d,iter+1,alpha)
        UA = X@UA
    return UA;

def UB_cal(d,alpha):
    UB = np.identity(d, dtype = 'complex_')
    for iter in range(d-1):
        Y = Ydj_alpha(d,iter+1,alpha)
        UB = UB@Y
    return UB

def Ud_ij(d,i,j):
    U = np.zeros([d,d], dtype = 'complex_')
    U[j,i] = 1;
    return U;
```
