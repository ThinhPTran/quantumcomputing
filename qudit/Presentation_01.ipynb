```python
import numpy as np

def Xdj_alpha(d,j,alpha):
    X = np.zeros((d,d), dtype = 'complex_')
    a = alpha[j];
    sn1 = 0;
    sn = 0;
    tmp = 0;
    for iter in range(j):
        tmp = tmp + alpha[iter]*np.conjugate(alpha[iter])
    sn1 = np.sqrt(tmp)
    sn = np.sqrt(tmp + alpha[j]*np.conjugate(alpha[j]))
    for iter in range(d):
        X[iter,iter]=1.0
    X[j-1,j-1]=a/sn
    X[j-1,j]=-sn1/sn
    X[j,j-1]=sn1/sn
    X[j,j]=np.conjugate(a)/sn
    return X;
```

```python
a12= np.array([[1/np.sqrt(2)],[1/np.sqrt(2)],[0]]);
alpha = np.array([1/np.sqrt(2),1/np.sqrt(2),0]);
```

```python
print(a12)
```

```python
print(alpha)
```

```python
X1=Xdj_alpha(3,1,alpha)
X2=Xdj_alpha(3,2,alpha)
```

```python
print(np.matmul(np.matmul(X2,X1),a12))
```

```python
import numpy as np

def Zd_theta(d,theta):
    Z = np.zeros((d,d), dtype = 'complex_')
    Z[d-1,d-1] = np.exp(theta*1j)
    return Z;
```

```python
print(Zd_theta(3,np.pi/2.0))
```

```python
print(np.exp(1j*np.pi/2.0))
```
