```python
import numpy as np

def X3j_alpha(j,alpha):
    X = np.zeros((3,3))
    a = alpha[j];
    sn1 = 0;
    sn = 0;
    tmp = 0;
    for iter in range(j):
        tmp = tmp + alpha[iter]*np.conjugate(alpha[iter])
    sn1 = np.sqrt(tmp)
    sn = np.sqrt(tmp + alpha[j]*np.conjugate(alpha[j]))
    for iter in range(3):
        X[iter,iter]=1.0
    X[j-1,j-1]=a/sn
    X[j-1,j]=-sn1/sn
    X[j,j-1]=sn1/sn
    X[j,j]=np.conjugate(a)/sn
    return X;
```

```python
a12= np.array([[1/np.sqrt(2)],[1/np.sqrt(2)],[0]]);
alpha = np.array([1/np.sqrt(2),1/np.sqrt(2),0]);
```

```python
print(a12)
```

```python
print(alpha)
```

```python
X1=X3j_alpha(1,alpha)
```

```python
X2=X3j_alpha(2,alpha)
```

```python
print(np.matmul(np.matmul(X2,X1),a12))
```
