```python
import numpy as np

def Xdj_alpha(d,j,alpha):
    X = np.zeros((d,d), dtype = 'complex_')
    a = alpha[j];
    sn1 = 0;
    sn = 0;
    tmp = 0;
    for iter in range(j):
        tmp = tmp + alpha[iter]*np.conjugate(alpha[iter])
    sn1 = np.sqrt(tmp)
    sn = np.sqrt(tmp + alpha[j]*np.conjugate(alpha[j]))
    for iter in range(d):
        X[iter,iter]=1.0
    X[j-1,j-1]=a/sn
    X[j-1,j]=-sn1/sn
    X[j,j-1]=np.conjugate(sn1/sn)
    X[j,j]=np.conjugate(a)/sn
    return X;
```

```python
import numpy as np

def Ydj_alpha(d,j,alpha):
    Y = np.zeros((d,d), dtype = 'complex_')
    a = alpha[j];
    sn1 = 0;
    sn = 0;
    tmp = 0;
    for iter in range(j):
        tmp = tmp + alpha[iter]*np.conjugate(alpha[iter])
    sn1 = np.sqrt(tmp)
    sn = np.sqrt(tmp + alpha[j]*np.conjugate(alpha[j]))
    for iter in range(d):
        Y[iter,iter]=1.0
    Y[j-1,j-1]=np.conjugate(a/(sn*sn))
    Y[j-1,j]=sn1/(sn*sn)
    Y[j,j-1]=-np.conjugate(sn1/(sn*sn))
    Y[j,j]=a/(sn*sn)
    return Y;
```
